```python
import nltk
from nltk.corpus.reader.plaintext import PlaintextCorpusReader
from nltk.probability import FreqDist
from collections import Counter
import math
import matplotlib.pyplot as plt
```

### How many total unique words are in the corpus?
* Alphanumeric tokens only.

```python
corpus = PlaintextCorpusReader('.', 'california-dams-article.txt')
tokens = nltk.word_tokenize(corpus.raw().strip())
len(set(tokens))
```

```python
alphaNumTokens = [t.lower() for t in tokens if t.isalnum()]
alphaNumTokensSet = set(alphaNumTokens)
len(alphaNumTokensSet)
```

2,081 unique words

### Taking the most common words, how many unique words represent half of the total words in the corpus?

```python
freqDist = FreqDist(alphaNumTokens)
counts = dict()
for word, count in freqDist.items():
    counts[word] = count

counts
```

```python
len(alphaNumTokens)
```

```python
wordCounter = Counter(alphaNumTokens).most_common()
wordCounter
```

The number of words in the whole corpus is 7,115. The most frequently occurring word is 'the' with 585 occurrences. So, there are no words that represent half of the total words in the corpus.

### Identify the 200 highest frequency words in this corpus.

```python
wordCounter[:200]
```

Most common words, not including stopwords:

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
top200 = []
for word, count in wordCounter:
    if word not in stop_words:
        top200.append((word, count))

top200
```

### Create a graph that shows the relative frequency of these 200 words.
* Includes stopwords

```python
from matplotlib.pyplot import figure

figure(figsize=(30, 4), dpi=80)

freqDist.plot(200, title="Relative Frequency, Top 200 Words", cumulative=True)
```

```python
x=[]
y=[]
for n, (word, count) in enumerate(wordCounter):
    x.append(math.log(n+1))
    y.append(math.log(count))
```

```python
plt.scatter(x, y, alpha=0.1)
plt.xlabel("Log Rank")
plt.ylabel("Log Frequency")
plt.title("Word Frequency vs Rank, Zipf's Law")
plt.show()
```

The rank metric is based on frequency. More frequently occurring words have a higher rank (lower rank number). For example, the word "the" has the highest rank, 1.
As rank decreases, words occur less frequently.
The color alpha was set to 0.1 to make it easier to see the frequency of higher ranked words compared to lower ranked words. There are many more words at the end of the tail than the beginning, as indicated by the darker blue. These words are lower in rank.
The points appear to form a very rough diagonal line with a negative slope. The direction holds for Zipf's Law, but ideally the diagonal would be straight without sloppy tails.
